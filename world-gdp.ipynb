- The function below cleans the population column removing the k's and M's in the column and converting them to their equivalent numerical formats

```python
def value_to_float(x):
    try:
        x = x.upper()
        if 'B' in x: 
            return float(x.replace('B', '')) * 10**9
        elif 'M' in x:
            return float(x.replace('M', '')) * 10**6
        elif 'K' in x:
            return float(x.replace('K', '')) * 10**3
        else:
            return float(x)
        return 0.0
    except Exception:
        return 0.0

population['population'] = population['population'].apply(value_to_float)
population.head()
```

- Pycountry could not match the alpha 2 code for **Timor-Leste** for some weird reason.
- The function below was supposed to generate the continent code from the country code.
- I had to resort to **option b** to get the country names

- This funmctions generates continent codes from the country codes

```python
def convert(row):
    country_code = pc.country_name_to_country_alpha2(row.country, cn_name_format = "default")
    conti_code = pc.country_alpha2_to_continent_code(country_code)
    return conti_code
```

```python
gdp_exp_pop["coninent"] = gdp_exp_pop.apply(convert, axis=1)
gdp_exp_pop.head()
```

```python
import plotly.io as pio
fig = px.scatter(df,x="gdp", 
                 y="LifeExpectancy",color="continent", 
                 animation_frame= "Year",animation_group="country",
                 labels={"LifeExpectancy":"Life Expectancy", "gdp":"GDP per Capita", "continent":"Continent"},
                 size="population",
                 hover_name="country", 
                 log_x=True, size_max=100, 
                 range_x=[200, 60000], 
                 range_y=[0,100],
                 color_discrete_sequence=px.colors.qualitative.G10).update_layout(
                          font_family="Georgia, serif",
                          font_color="black",
                          legend_title_font_color="green", paper_bgcolor='rgba(0,0,0,0)',
                          plot_bgcolor='rgba(0,0,0,0)').update_xaxes(zeroline=False, showgrid=False,
                                                                     title_font_family="Georgia, serif").update_yaxes(
                          zeroline=False, showgrid=False)
fig.update_layout(title_text="<b>Exploring GDP, LifeExpectancy and Population (1900-2023)<b>", title_x=0.5,
                  title_font_size = 16,
                        font_family="Georgia, serif", title_font_family="Georgia, serif", 
                        title_font_color="black", legend_title_font_color="green",
                        yaxis = dict(tickfont = dict(size=14)),
                        xaxis = dict(tickfont = dict(size=20))
                        ,sliders = [dict(font={'size': 17})] 
                 )
fig.show(pio="browser")
```
